```python
remember = NE['remember'] + BF['remember'] + UT['remember']
understand = NE['understand'] + BF['understand'] + UT['understand']
apply = NE['apply'] + BF['apply'] + UT['apply']
analyze = NE['analyze'] + BF['analyze'] + UT['analyze']
evaluate = NE['evaluate'] + BF['evaluate'] + UT['evaluate']
create = NE['create'] + BF['create'] + UT['create']
```

```python
remember = sorted(set(remember))
understand = sorted(set(understand))
apply = sorted(set(apply))
analyze = sorted(set(analyze))
evaluate = sorted(set(evaluate))
create = sorted(set(create))
```

```python
print(evaluate)
```

```python
all_words = remember + understand + apply  + analyze + evaluate + create

word_overlap = {
    word: sum(1 for s in [remember, understand, apply, analyze, evaluate, create] if word in s)
    for word in all_words
}

multiple_overlap_words = {word: count for word, count in word_overlap.items() if count > 1}
once_words = {word: count for word, count in word_overlap.items() if count == 1}

print(len(multiple_overlap_words))
print(len(once_words))


print("여러 집합에서 등장하는 단어와 등장 횟수:")
for word, count in multiple_overlap_words.items():
    print(f"{word}: {count}개 집합에서 등장")
```
